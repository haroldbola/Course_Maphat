### Instructions
```You are to build two models:```

```discriminator:```
- ```Input: 2 length vector```
- ```Output: p, p between 0 and 1, the probability of the input vector to be normally distributed```
    
```generator (read the further instructions below):```
- ```Input: 20 random numbers, noise```
- ```Output: a 2 length vector```

```When training the discriminator, the input is to be made out of two segments: vectors which are actually from the desired distribution and vectors generated by the generator. Note that you will probably want to generate these vectors with the newest version of the generator before each training session.```

```When training the generator, the input should be random noise, and the predictions should be the probability that you want the discriminator to predict, i.e, vector of ones. When training the generator, you will have to regard the discriminator as a layer, and use it on the generated vectors. To do so you will have to freeze the weights in the discriminator, so when you train the generator, the discriminator will be fixed. This means that you will have to create a model that is the discriminator, composed on the generator ( discriminator(generator(x)) ), and train it (when the discriminator is fixed).```
