## Reading and preprocessing the data you created
create a DataFrame with 2 columns:
1. The exact path of each image
2. The label of each image

Split this data into train and test sets

```python
# paths
path_gal = './data/gal/'
path_philip = './data/philip/'

gal_files = [(path_gal+file) for file in os.listdir(path_gal)] 
philip_files = [(path_philip+file) for file in os.listdir(path_philip)] 
```

```python
# prepare df
labels = np.concatenate((np.full(len(gal_files),'gal'), np.full(len(philip_files),'philip')))
all_paths = np.concatenate((gal_files,philip_files))

df = pd.DataFrame({'path': all_paths, 'label': labels})
df.head(2)
```

```python
## Alternative way for performance the actions above:
# df = pd.DataFrame(columns=['image_path', 'label'])
# for folder in listdir('./data'):
#     for im in listdir(path.join('./data', folder)):
#         df = df.append({'image_path': path.join('./data', folder, im),
#                   'label': folder}, ignore_index=True)
```

```python
df_train, df_test, target_train, target_test = train_test_split(df['path'], df['label'], test_size= 0.2)
```

Now use iterate through the dataframes and read the images corresponding to each row (using keras.preprocessing.image). Save the images as numpy arrays, and once again use utils.preprocess_input to preprocess the arrays.
Save the labels in 2 differet ways:
1. Onehot encoding - for using a NN
2. Serial number of the corresponding class (one number per label) - for using an SVM

```python
def my_load_img(df):
    # load all images from df by path 
    images= []
    target_size = (224,224)
    for i in range(len(df)):
        img = load_img(df.iloc[i], target_size= target_size)
        img = image.img_to_array(img)
        images.append(utils.preprocess_input(img))
    return np.asanyarray(images)
```

```python
# load train & test images
train_images = my_load_img(df_train)
test_images = my_load_img(df_test)
```

```python
# prepare labels for SVM
svm_target_train = [0 if label=='gal' else 1 for label in target_train ]
svm_target_test = [0 if label=='gal' else 1 for label in target_test]
```

```python
# prepare labels for NN
labels_num= 2
nn_target_train = np.array(np.eye(labels_num))[svm_target_train]
nn_target_test = np.array(np.eye(labels_num))[svm_target_test]
```

Create an instance of keras.preprocessing.image.ImageDataGenerator, which will define how you will create augmentations of each original image you've created. Chose all the parameters on your own.

```python
# Image Data Generator
generator= keras.preprocessing.image.ImageDataGenerator(rescale=1./255, featurewise_center= False, samplewise_center=True,
                                                        featurewise_std_normalization=False, samplewise_std_normalization=False, rotation_range=40, width_shift_range=0.0, height_shift_range=0.0, brightness_range=None, zoom_range=[0.5,1.5], validation_split=0.0)
```

# Finally - transfer learning. 
### This should be done with the keras functional API (not Sequential)
1. Create an instance of VGGFace with the model of your choice. 
2. Examine the model's architechture using .summary(). Understand what it means to replace the last 2 dense layers (including the final softmax layer). 
3. Use .get_layer() to retrive the last layer which you want to keep in your new NN.
4. Create the 2 new Dense layers which continue the previous pretrained layers (the last layer should have a softmax activation).
5. Create a new model which the input of the original model as input, and outputs the new dense-softmax layer. You can use .input on the old model for this.
6. Freeze all of the layers except the last 2 using .layers on the new model, and .trainable = False. This will stop you from training those layers.
7. Compile the model ('adam' worked ok for us as an optimizer).

```python
# 1
vgg = VGGFace(model='vgg16')
```

```python
# 2
vgg.summary()
```

```python
# vgg.layers.pop()
```

```python
# 3- take last layers of VGG
partial_model = Model(vgg.input, vgg.get_layer('fc7/relu').output)
```

```python
# 4- new Dense layers
my_dense_1 = Dense(1000, activation='tanh')(partial_model.output)
my_dense_2 = Dense(labels_num, activation='softmax')(my_dense_1)
```

```python
# # Better Way:
# # 3- take last layers of VGG
# last_layer_to_keep = vgg.get_layer(name='fc7/relu')
# my_dense_1 = Dense(1000, activation='tanh')(last_layer_to_keep.output)
# my_dense_2 = Dense(labels_num, activation='softmax')(my_dense_1)
```

```python
new_model = Model(vgg.input, my_dense_2)

# freeze the oroginal layers
for layer in new_model.layers[:-2]:
    layer.trainable = False
    
new_model.compile(optimizer='rmsprop', loss= 'categorical_crossentropy', metrics=['acc']) ### MAYBE binary_crossentropy??
new_model.summary()
```

Now you're ready to train the model:
1. Use .fit_generator() and not .fit(), since you'll be using the augmentor you created.
2. Use .flow() on the instance of ImageDataGenerator as the first input. 
3. Choose a combination on batch_size(within .flow) and steps_per_epoch which will create a total number of images that you want per each epoch. 
4. Use validation data which isn't augmented.

How well is it doing? We got an accuracy of 90%-95% percent (validation data) on photos on the 2 of us (just 2 classes).

```python
new_model.fit_generator(generator=generator.flow(train_images, nn_target_train, batch_size= 12), steps_per_epoch=20, epochs=36, verbose=1, validation_data=(test_images, nn_target_test))
```

We got accuracy of 98% on validation (test data)! [about 36 ephocs]
